Determine barbasi vs erdos reiny avg degree and some proporties

```python
from os.path import join, isdir
from os import listdir
import json
import networkx as nx
import itertools
import matplotlib.pyplot as plt
from networkx.generators.random_graphs import erdos_renyi_graph, barabasi_albert_graph
import numpy as np
%matplotlib inline
```

```python
EG = erdos_renyi_graph(30, p=0.20)
np.mean(list(dict(nx.degree(EG)).values()))
```

```python
2 * len(EG.edges()) / len(EG.nodes())
```

```python
ns = list(range(30, 41, 1))
ps = np.arange(0.05, 1.05, 0.05)
```

```python
def _community_average_degree(g, community_vertices: list):
    """
    Computes community's average degree score, given a list of vertices.
    The computation is done over the sub network induced from the community's vertices.
    Based on internal consistency only.
    Communities ranked at the boundaries are anomalous (normal)? which way to look?
    """

    comm_edges = nx.subgraph(g, community_vertices).edges()
    return 2 * len(comm_edges) / len(community_vertices)
```

```python
n= 34
p = 0.2
for _ in range(50):

    BG = barabasi_albert_graph(n=n, m=1)
    avg_b = np.mean(list(dict(nx.degree(BG)).values()))
    p = 0.2
    EG = erdos_renyi_graph(n, p=p)
    avg_e = np.mean(list(dict(nx.degree(EG)).values()))

    print(f'{n} Nodes: avg. barabsi: {avg_b:.3f}, avg. erdos-reiny (p={p:.2f}): {avg_e:.3f}')
```

```python
for n in ns:
    for p in ps:

        BG = barabasi_albert_graph(n, m=1)
        avg_b = np.mean(list(dict(nx.degree(BG)).values()))
        
        cum_avg_e = 0
        for _ in range(50):
            EG = erdos_renyi_graph(n, p=p)
            cum_avg_e += np.mean(list(dict(nx.degree(EG)).values()))
        cum_avg_e/=50
        
        print(f'{n} Nodes: avg. barabsi: {avg_b:.3f}, avg. erdos-reiny (p={p:.2f}): {cum_avg_e:.3f}')
        
```
